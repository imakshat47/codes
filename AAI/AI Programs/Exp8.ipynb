## **Theory**

**Reasoning Under Uncertainty using Bayesian Learning**

1. Introduction:
Bayesian learning is a probabilistic approach to reasoning under uncertainty, where knowledge about an event is updated as new evidence is observed. It is based on Bayes’ Theorem, which relates conditional and marginal probabilities of events.

2. Bayes’ Theorem:
Bayes’ Theorem helps in computing the posterior probability P(H|E), i.e., the probability of hypothesis H given evidence E.

3. Example (Wet Grass Problem):
We consider a scenario with three events: Rain, Sprinkler, and Wet Grass. Wet grass can result from either rain or the sprinkler being on. Given the grass is wet, we use Bayes’ Theorem to compute the probability it rained.


## **Code & Output**

```python
# Define probabilities
P_rain = 0.2                    # P(Rain)
P_sprinkler_given_rain = 0.01  # P(Sprinkler|Rain)
P_sprinkler_given_no_rain = 0.4
P_wet_grass_given_rain_sprinkler = 0.99
P_wet_grass_given_rain_no_sprinkler = 0.9
P_wet_grass_given_no_rain_sprinkler = 0.8
P_wet_grass_given_no_rain_no_sprinkler = 0.0

# Use total probability theorem to calculate P(WetGrass)
P_no_rain = 1 - P_rain
P_sprinkler_and_rain = P_rain * P_sprinkler_given_rain
P_no_sprinkler_and_rain = P_rain * (1 - P_sprinkler_given_rain)
P_sprinkler_and_no_rain = P_no_rain * P_sprinkler_given_no_rain
P_no_sprinkler_and_no_rain = P_no_rain * (1 - P_sprinkler_given_no_rain)

# Total probability of wet grass
P_wet_grass = (
    P_sprinkler_and_rain * P_wet_grass_given_rain_sprinkler +
    P_no_sprinkler_and_rain * P_wet_grass_given_rain_no_sprinkler +
    P_sprinkler_and_no_rain * P_wet_grass_given_no_rain_sprinkler +
    P_no_sprinkler_and_no_rain * P_wet_grass_given_no_rain_no_sprinkler
)

# Bayes Theorem: P(Rain | Wet Grass)
P_rain_and_wet_grass = (
    P_sprinkler_and_rain * P_wet_grass_given_rain_sprinkler +
    P_no_sprinkler_and_rain * P_wet_grass_given_rain_no_sprinkler
)
P_rain_given_wet_grass = P_rain_and_wet_grass / P_wet_grass

print(f"P(Wet Grass) = {P_wet_grass:.3f}")
print(f"P(Rain | Wet Grass) = {P_rain_given_wet_grass:.3f}")
```
