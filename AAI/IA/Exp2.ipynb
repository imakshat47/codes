**Google Colab**

Google Colab is a free, cloud-based Jupyter Notebook environment that allows you to write and execute Python code in your browser, with zero setup and free access to GPUs.

How to use:
1. Go to https://colab.research.google.com
2. Sign in with Google → Click “New Notebook”
3. Start coding in the code cells, add text in text (Markdown) cells.
4. Use Shift+Enter to run code or Ctrl+M+B to add a new cell.

Advantages:
 (like Google Docs)

*   No setup needed
*   Easily share notebooks
*   Run on GPU/TPU


**Jupyter Notebook**

Jupyter Notebook is a web-based interactive environment to write, run, and visualize Python code locally.

Installation: pip install notebook

Launch: jupyter notebook

It will open in your browser at http://localhost:8888/. You can create .ipynb notebooks and write code in cells just like in Colab.

Advantages:
*   Run offline
*   Good for data analysis & machine learning
*   Visualize plots and outputs inline
