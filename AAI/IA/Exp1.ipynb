## **Theory**

1. Variables and Data Types
Variables are used to store data values and are created when first assigned.
Python supports various data types like int, float, str, and bool.

2. Operators
Operators are special symbols used to perform operations on variables and values.
Common operators include arithmetic (+, -), comparison (==, <) and logical (and, or).

3. Control Flow
Control flow lets you execute code conditionally using if, elif, and else statements.
Loops (for, while) allow repeating a block of code multiple times based on conditions.

4. Functions
Functions are reusable blocks of code that perform a specific task and can return results.
They are defined using the def keyword and can accept parameters.

5. Data Structures

 5.1 List
Lists are ordered, mutable collections used to store multiple items in a single variable.
You can add, remove, or access elements using index positions.

 5.2 Tuple
Tuples are ordered and immutable collections, meaning their values cannot be changed.
They are useful for fixed data like coordinates or days of the week.

 5.3 Set
Sets are unordered collections of unique elements, useful for eliminating duplicates.
You can add or remove items, but indexing is not allowed.

 5.4 Dictionary
Dictionaries store data in key-value pairs, allowing fast lookups by key.
They are mutable and ideal for structured data like user profiles.\

## **Code & Output**

```python
# Introduction to Python Programming

# 1. VARIABLES AND DATA TYPES
print("1. VARIABLES AND DATA TYPES")
name = "ABC"
age = 25
height = 5.9
is_student = True

print(f"Name: {name}, Age: {age}, Height: {height}, Student: {is_student}\n")

# 2. OPERATORS
print("2. OPERATORS")
a = 10
b = 3
print(f"Addition: {a + b}")
print(f"Subtraction: {a - b}")
print(f"Multiplication: {a * b}")
print(f"Division: {a / b}")
print(f"Modulus: {a % b}")
print(f"Exponent: {a ** b}")
print(f"Floor Division: {a // b}\n")

# 3. CONTROL FLOW
print("3. CONTROL FLOW")

# if-elif-else
num = 7
if num > 0:
    print("Positive number")
elif num == 0:
    print("Zero")
else:
    print("Negative number")

# for loop
print("\nFor Loop (0 to 4):")
for i in range(5):
    print(i, end=' ')
print()

# while loop
print("While Loop (5 to 1):")
i = 5
while i > 0:
    print(i, end=' ')
    i -= 1
print("\n")

# 4. FUNCTIONS
print("4. FUNCTIONS")
def greet(name):
    return f"Hello, {name}!"

print(greet("ABC"))

def add(x, y):
    return x + y

print(f"Addition: {add(5, 7)}\n")

# 5. DATA STRUCTURES

# 5.1 LIST
print("5.1 LIST")
fruits = ["apple", "banana", "cherry"]
fruits.append("orange")
print("Fruits:", fruits)
print("First fruit:", fruits[0])
fruits.remove("banana")
print("After removal:", fruits, "\n")

# 5.2 TUPLE
print("5.2 TUPLE")
coordinates = (10, 20)
print("Coordinates:", coordinates)
print("X:", coordinates[0], "\n")

# 5.3 SET
print("5.3 SET")
unique_numbers = {1, 2, 2, 3, 4}
unique_numbers.add(5)
print("Set elements:", unique_numbers)
unique_numbers.remove(3)
print("After removal:", unique_numbers, "\n")

# 5.4 DICTIONARY
print("5.4 DICTIONARY")
student = {
    "name": "ABC",
    "age": 25,
    "grade": "A"
}
print("Student Name:", student["name"])
student["age"] = 26
print("Updated Age:", student["age"])
print("Keys:", student.keys())
print("Values:", student.values())
```
