## **Theory**

**Introduction to Machine Learning and Python Libraries**

**1. What is Machine Learning?**

Machine Learning (ML) is a branch of artificial intelligence (AI) where machines learn patterns from data to make predictions or decisions without being explicitly programmed. It is categorized into:


	•	Supervised Learning: Learn from labeled data (e.g., Linear Regression).
	•	Unsupervised Learning: Find patterns in unlabeled data (e.g., Clustering).
	•	Reinforcement Learning: Learn from interaction with the environment.

**2. Pandas:**

Pandas is a powerful library for data manipulation and analysis. It allows easy handling of tabular data using DataFrames and Series. Common uses include reading datasets, cleaning data, and performing descriptive analysis.

**3. NumPy:**

NumPy provides support for numerical computations and is heavily used in ML for operations on arrays and matrices. It powers many ML algorithms that require linear algebra or matrix operations.

**4. Matplotlib:**

Matplotlib is a plotting library used for data visualization. It helps in understanding the data trends through graphs like line plots, scatter plots, and bar charts.


## **Code & Output**

```python
# Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# Create a smaller dataset (small grid: 5 data points)
data = {
    'Hours_Studied': [1, 2, 3, 4, 5],
    'Exam_Score': [35, 45, 50, 60, 70]
}

# Load into a DataFrame
df = pd.DataFrame(data)

# Display the dataset
print("Dataset:\n", df)

# Plotting the data
plt.scatter(df['Hours_Studied'], df['Exam_Score'], color='blue')
plt.title('Hours vs Score (Small Grid)')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.grid(True)
plt.show()

# Prepare data for training
X = df[['Hours_Studied']]  # Features
y = df['Exam_Score']       # Target

# Split into training and testing sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

# Apply Linear Regression
model = LinearRegression()
model.fit(X_train, y_train)

# Predict
y_pred = model.predict(X_test)

# Evaluate
mse = mean_squared_error(y_test, y_pred)
print("\nMean Squared Error:", mse)
print("Predicted scores:", y_pred)

# Plot the regression line
plt.scatter(X, y, color='blue')
plt.plot(X, model.predict(X), color='red')
plt.title('Regression Line (Small Grid)')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.grid(True)
plt.show()
```
